## Loading the iris dataset into scikit-learn

```python
# import load_iris function from dataset module
from sklearn.datasets import load_iris
```

```python
# save "bunch" object containing iris dataset and its attributes
iris = load_iris()
type(iris)
```

```python
# print the iris data
print(iris.data)
```

## Machine learning terminology

- Each row is an __observation__ (also known as: sample, example, instance, record)
- Each column is a __feature__ (also known as: predictor, attribute, independent variable, input, regressor, covariate)

```python
# print the names of the four features
print(iris.feature_names)
```

```python
# print integers representing the species of each observation
print(iris.target)
```

```python
# print the encoding scheme for species: 0 = setosa, 1 = versicolor, 2 = virginica
print(iris.target_names)
```

- Each value we are predicting is a __responce__ (also known as: target, outcome, label, dependent variable)

These are two types of machine learning problem:
- __Classification__ is supervised learning in which the response is categorical (
- __Regression__ is supervised learning on which the response is ordered and continuous

## Requirements for working with data in scikit-learn

1. Features and response are __separate objects__
1. Features and response should be __numeric__
1. Features and response should be __NumPy arrays__
1. Features and responses should have __specific shapes__

```python
# check the types of the features and response (both should be numpy arrays - 'ndarray')
print(type(iris.data))
print(type(iris.target))
```

```python
# check the shape of the features (first dimension = number of observations, second dimension = number of features)
print(iris.data.shape)
```

```python
# check the shape of the response (single dimension matching the number of observations)
print(iris.target.shape)
```

```python
# store feature matrix in "X" - X is the scikit-learn convention for a feature
X = iris.data

# store response vector in "y" - y is the scikit-learn convention for a response
y = iris.target
```

- The 'X', is capitalised to indicate that it is a matrix

- The 'y', is lower case to indicate that it is a vector
